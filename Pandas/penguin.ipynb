```python
clean_penguins = penguins.dropna().reset_index(drop=True)
clean_penguins
```

```python
# Clean by fill na.
top_penguins = penguins.head(10)
bill_length_mean = top_penguins['bill_length_mm'].mean()
top_penguins = top_penguins['bill_length_mm'].fillna(value=bill_length_mean)
top_penguins
```

```python
penguins.dropna().sort_values(['species', 'island', 'body_mass_g']).head(10)
```

```python
# Unique 
penguin_species = list(penguins['species'].unique())
penguin_island = list(penguins['island'].unique())
print(f"Species: {penguin_species}\nIsland: {penguin_island}")
```

```python
# Independant Variables 
inde_var_count = penguins[['species', 'island']].value_counts().reset_index()
inde_var_count.columns = ['species', 'island', 'count']
inde_var_count
```

```python
penguins_bodymass_desc = penguins.groupby(['species', 'island'])['body_mass_g'] \
  .agg(['min', 'max', 'mean', 'median', 'std']) \
	.reset_index()
penguins_bodymass_desc.to_csv("Penguin_BodyMass_Description.csv")
```
